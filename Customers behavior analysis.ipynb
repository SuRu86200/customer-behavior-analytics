```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create age_group column
labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] =pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_freaquency_day
frequency_mapping = {'Fortnightly' : 14,
                      'Weekly' : 7,
                      'Monthly' : 30,
                      'Quarterly' : 90,
                      'Bi-Weekly' : 14,
                      'Annually' : 365,
                      'Everey 3 Months' : 90}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
                      
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Connection Details
username = 'root'
password = 'Surendra@9123'
host     = 'localhost'
port     = '3306'
database = 'customer_behavier'

# 2. Create Engine (Updated to use mysql-connector)
engine = create_engine(f"mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}")

# 3. Write DataFrame to MySQL
table_name = 'customers_data'
df.to_sql(table_name, con=engine, if_exists='replace', index=False)
print(f"Success: Data written to table '{table_name}'")

# 4. Read back (Fixed the table name here)
# We query the TABLE name 'customers_data', not the database name.
df_from_db = pd.read_sql(f"SELECT * FROM {table_name} LIMIT 5;", con=engine)

print(df_from_db)
```
